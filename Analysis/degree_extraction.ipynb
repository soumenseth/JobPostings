```python
nlp = spacy.load('en_core_web_sm') 
matcher = PhraseMatcher(nlp.vocab)
```

```python
patterns = ["Bachelor's", "Bachelor’s Degree", "Bachelors", "Master's", "MS", "BTech", "phd", "Master’s Degree", "Masters", "PhD"]
for pattern in patterns:
    matcher.add('DEGREE', [nlp(pattern)])
```

```python
linkedid_data_path = 'Data/linked_in.json'
linkedid_parsed_data_path = 'Data/linked_in_parsed.json'
```

```python
jobs_parsed = read_json(linkedid_parsed_data_path)
jobs = read_json(linkedid_data_path)
```

```python
degrees = []
for job in jobs[:100]:
    doc = nlp(job)
    matches = matcher(doc)
    if len(matches):
        for match_id, start, end in matches:
            string_id = nlp.vocab.strings[match_id]  # Get string representation
            span = doc[start:end]  # The matched span
            degrees.append({
                'match_id': match_id,
                'string_id': string_id,
                'start': start,
                'end': end,
                'text': span.text
            })
```
